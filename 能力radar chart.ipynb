```python
import matplotlib.pyplot as plt
import numpy as np

# Set the values for each skill
detail_oriented = 5
research_experience = 5
communication_skills = 3
recruiting_experience = 5
working_with_older_adults = 5
independent_research = 5
evening_or_weekend_work = 5
neuroimaging_experience = 4
statistical_analysis = 4
programming_skills = 4

# Set the names of the skills
skills = ['Detail-oriented', 'Research experience', 'Communication skills',
          'Recruiting experience', 'Working with older adults', 'Independent research',
          'Evening/weekend work', 'Neuroimaging experience', 'Statistical analysis',
          'Programming skills']

# Set the values for each skill in a list
values = [detail_oriented, research_experience, communication_skills,
          recruiting_experience, working_with_older_adults, independent_research,
          evening_or_weekend_work, neuroimaging_experience, statistical_analysis,
          programming_skills]

# Set the number of skills and the angle at which to start the first skill
num_skills = len(skills)
angle_start = np.pi/2

# Calculate the angles for each skill
angles = [angle_start + i*2*np.pi/num_skills for i in range(num_skills)]
angles += angles[:1]

# Add the first value to the end of the values list to close the polygon
values += values[:1]

# Create the plot
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, values, 'o-', linewidth=2)
ax.fill(angles, values, alpha=0.25)
ax.set_xticks(angles[:-1])  # Set tick locations without the last angle
ax.set_xticklabels(skills)  # Set tick labels using the skill names
ax.set_title('Skills Radar Chart', fontsize=14)
ax.grid(True)

# Show the plot
plt.show()
```
